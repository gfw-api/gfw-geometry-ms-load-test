```python
import pandas as pd
```

```python
df = pd.read_csv('logs/results.csv')

# convert string percentage value to float
df.docker_cpu_pct = df.apply(lambda row: float(row['docker_cpu_pct'].strip('%')), axis=1)
df.head()
```

```python
%matplotlib inline

# get into a format so that pandas will plot all on the same graph
ax = df.pivot(index="time_step", columns="test_name", values="docker_cpu_pct").plot(figsize=(15,8))
ax.set(ylabel="docker stats CPU (Percent)")
```
